```python
from matplotlib import pyplot as plt
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples=1000, random_state=42, noise=0.1)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=10, c=y)
```
