##Generate Datasets : Moons

```python
from matplotlib import pyplot as plt
from sklearn.datasets import make_moons
```

```python
X, y = make_moons(n_samples=1000, random_state=42, noise=0.1)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=10, c=y)
```

##Generate Datasets : Blobs

https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html#sklearn.datasets.make_blobs

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=10, centers=1, n_features=2, random_state=0)
```

```python
# Generate sample data
X, y = make_blobs(n_samples=4000, centers=5, n_features=2)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=5, c=y)
```

```python
from sklearn.datasets import make_gaussian_quantiles
X, y = make_gaussian_quantiles(cov=2.0, n_samples=200, n_features=2, n_classes=2, random_state=1)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=15, c=y)
```

```python
from sklearn.datasets import make_hastie_10_2
X, y = make_hastie_10_2(n_samples=120, random_state=1)
```

```python
plt.scatter(X[:, 0], X[:, 1], s=10, c=y)
```
